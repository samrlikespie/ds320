```python
import difflib

# standardize the Address and Location_Block features
scrimes['Location_Block_std'] = scrimes['Location_Block'].str.lower().str.replace(",", "").str.replace(".", "").str.replace(" ", "")
colleges['Address_std'] = colleges['ADDRESS'].str.lower().str.replace(",", "").str.replace(".", "").str.replace(" ", "")

# use difflib to get a ratio of similarity between the standardized features
colleges['Address_match'] = colleges['Address_std'].apply(lambda x: scrimes['Location_Block_std'].apply(lambda y: difflib.SequenceMatcher(None, x, y).ratio()).max())

# merge the data frames based on matching addresses
merged = pd.merge(scrimes, colleges, how='left', left_on='Location_Block_std', right_on='Address_std')

# display the result
#print(merged[['Address', 'Crime_Type', 'College_Name']])
```
